## 몬테카를로 구현

### 그리드월드(분포 모델 구현)

### 상태가치함수

```python
import numpy as np
import gridworld_render as render_helper

class GridWorld:
    
    def __init__(self):
        self.action_space = [0, 1, 2, 3] # action index
        
        self.action_meaning = {0:'업', 
                               1:'다운',
                               2:'왼쪽',
                               3:'오른쪽'
        } # name of action
        
        self.reward_map = np.array(
        [[0, 0, 0, 1],
         [0, None, 0, -1],
         [0, 0, 0, 0]]
        ) # reward in space
        
        self.goal_state = (0, 3) # 목표
        self.wall_state = (1, 1) # 벽 좌표
        self.start_state = (2, 0) # 시작 좌표
        self.agent_state = self.start_state # 에이전트 초기 좌표
        
    @property
    def height(self):
        return len(self.reward_map)

    @property
    def width(self):
        return len(self.reward_map[0])

    @property
    def shape(self):
        return self.reward_map.shape
    
    def actions(self):
        return self.action_space
    
    def states(self): # iterable 인스턴스로 생성
        for h in range(self.height):
            for w in range(self.width):
                yield (h, w)
                
    def next_state(self, state, action):
        # 이동위치 계산
        action_move_map = [(-1, 0), (1, 0), (0, -1), (0, 1)] # 업, 다운, 왼쪽, 오른쪽을 행렬 인덱스에서 표현
        move = action_move_map[action] # 액션 인덱스에 따라 움직임 설정
        next_state = (state[0] + move[0], state[1] + move[1])
        ny, nx = next_state
        
        if nx < 0 or nx >= self.width or ny < 0 or ny >= self.height:
            next_state = state
        elif next_state == self.wall_state:
            next_state = state
        
        return next_state
    
    def reward(self, state, action, next_state): # r(s, a, s')
        return self.reward_map[next_state]

    def render_v(self, v=None, policy=None, print_value=True): # V, pi 전달하여 상태공간 시각화 함수
        renderer = render_helper.Renderer(self.reward_map, self.goal_state,
                                          self.wall_state)
        renderer.render_v(v, policy, print_value)

    def render_q(self, q=None, print_value=True): # 이번에는 안쓰임
        renderer = render_helper.Renderer(self.reward_map, self.goal_state,
                                          self.wall_state)
        renderer.render_q(q, print_value)
        
    ## 2024.06.16 5장 추가 되는 부분 
    def step(self, action): # 에이전트가 행동하도록 하여 샘플 데이터를 얻도록 함
        state = self.agent_state
        next_state = self.next_state(state, action)
        reward = self.reward(state, action, next_state)
        done = (next_state == self.goal_state)

        self.agent_state = next_state
        return next_state, reward, done
    
    def reset(self): # 환경을 초기 상태로 재설정하는 메서드
        self.agent_state = self.start_state
        return self.agent_state
```

```python
from collections import defaultdict

class RandomAgent:
    def __init__(self):
        self.gamma = 0.9 # 할인율
        self.action_size = 4 # 행동 수

        random_actions = {0: 0.25, 1: 0.25, 2: 0.25, 3: 0.25} # 무작위 행동 확률 분포
        self.pi = defaultdict(lambda: random_actions) # 정책 설정
        self.V = defaultdict(lambda: 0) # 가치 함수
        self.cnts = defaultdict(lambda: 0) # 증분 방식으로 수익의 평균 구할 때 사용
        self.memory = [] # 실제 행동하여 얻은 경험 데이터(상태,행동,보상) 담기

    def get_action(self, state): # state에서 수행할 수 있는 행동을 하나 소환
        action_probs = self.pi[state]
        actions = list(action_probs.keys())
        probs = list(action_probs.values())
        return np.random.choice(actions, p=probs) # prob의 확률 분포에 따라 행동을 한 개씩 샘플링하는 코드

    def add(self, state, action, reward): # 수행한 행동과 보상을 기록
        data = (state, action, reward) # (상태,행동,보상) 튜플 데이터
        self.memory.append(data) # 데이터 memory 추가
        # 마지막 상태는 memory에 저장되지 않음, 목표 지점의 가치 함수는 항상 0
        # 갱신 필요가 없으므로 굳이 저장하지 않음
        
    def reset(self):
        self.memory.clear() 

    # 몬테카를로법 계산 수행
    def eval(self):
        G = 0 # 수익 0 초기화
        for data in reversed(self.memory): # a>b>c 수행하면, 수익은 c>b>a 역방향으로 계산
            state, action, reward = data
            G = self.gamma * G + reward
            self.cnts[state] += 1 # 증분 방식 평균 계산
            self.V[state] += (G - self.V[state]) / self.cnts[state]
```

### 코멘트 : EVAL, reversed : NAMING 수정해보기

```python
env = GridWorld()
```

```python
agent = RandomAgent()
```

```python
episodes = 1000 # 에피소드 1000번 수행
for episode in range(episodes):
    state = env.reset()
    agent.reset()

    while True:
        action = agent.get_action(state) # 행동 선택
        next_state, reward, done = env.step(action) # 행동 수행

        agent.add(state, action, reward) # 행동에 따른 (상태, 행동, 보상) 데이터 저장
        if done: # 목표 도달 시
            agent.eval() # 몬테카를로법 카치 함수 갱신
            break       # 다음 에피소드 수행

        state = next_state
```

## 행동 가치 함수 

```python
# 탐욕화 : 가치 함수를 최대화하여 개선 부분
# 최대화되도록 확률분포를 반환
# Q 함수 값이 가장 큰 행동만 취하게끔 확률 분포 셋팅
def greedy_probs(Q, state, epsilon=0, action_size=4):
    qs = [Q[(state, action)] for action in range(action_size)]
    max_action = np.argmax(qs)

    # 1번 개선
    base_prob = epsilon / action_size
    action_probs = {action: base_prob for action in range(action_size)}  #{0: ε/4, 1: ε/4, 2: ε/4, 3: ε/4}
    action_probs[max_action] += (1 - epsilon)
    return action_probs

# 정책 제어
# randomagent 유사함
class McAgent:
    def __init__(self):
        self.gamma = 0.9
        self.epsilon = 0.1
        self.alpha = 0.1
        self.action_size = 4

        random_actions = {0: 0.25, 1: 0.25, 2: 0.25, 3: 0.25}
        self.pi = defaultdict(lambda: random_actions)
        self.Q = defaultdict(lambda: 0) # 차이점 :  randomagent 에서 self.V임
        self.memory = []

    def get_action(self, state):
        action_probs = self.pi[state]
        actions = list(action_probs.keys())
        probs = list(action_probs.values())
        return np.random.choice(actions, p=probs)

    def add(self, state, action, reward):
        data = (state, action, reward)
        self.memory.append(data)

    def reset(self):
        self.memory.clear()

    def update(self): # self.Q 갱신하는 부분
        G = 0
        for data in reversed(self.memory):
            state, action, reward = data
            G = self.gamma * G + reward
            key = (state, action) # self.Q는 데이터가 상태, 행동
            self.Q[key] += (G - self.Q[key]) * self.alpha #2번 개선
            self.pi[state] = greedy_probs(self.Q, state, self.epsilon)
```

```python
env = GridWorld()
agent = McAgent()

episodes = 10000
for episode in range(episodes):
    state = env.reset()
    agent.reset()

    while True:
        action = agent.get_action(state)
        next_state, reward, done = env.step(action)

        agent.add(state, action, reward)
        if done:
            agent.update()
            break

        state = next_state

env.render_q(agent.Q)
```
